```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make sure your data is loaded
# df = pd.read_csv('your_data.csv')

# Now run the visualization code
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df['math score'], kde=True)
plt.title('Math Score Distribution')

plt.subplot(1, 3, 2)
sns.histplot(df['reading score'], kde=True)
plt.title('Reading Score Distribution')

plt.subplot(1, 3, 3)
sns.histplot(df['writing score'], kde=True)
plt.title('Writing Score Distribution')

plt.tight_layout()
plt.show()
```

```python
print("=== AVERAGE SCORES ===")
avg_scores = df[['math score', 'reading score', 'writing score']].mean()
print(avg_scores)

# Score trends by gender
print("\n=== SCORES BY GENDER ===")
gender_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
print(gender_scores)

# Score trends by race/ethnicity
print("\n=== SCORES BY RACE/ETHNICITY ===")
race_scores = df.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean()
print(race_scores)

# Score trends by parental education
print("\n=== SCORES BY PARENTAL EDUCATION ===")
parent_scores = df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
print(parent_scores.sort_values('math score', ascending=False))
```

```python
print("=== TEST PREPARATION IMPACT ===")
prep_impact = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
print(prep_impact)

# Lunch type impact
print("\n=== LUNCH TYPE IMPACT ===")
lunch_impact = df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()
print(lunch_impact)

# Strongest predictors - correlation analysis
print("\n=== CORRELATION ANALYSIS ===")
# Convert categorical to numeric for correlation
df_corr = df.copy()
df_corr['test_prep_num'] = df_corr['test preparation course'].map({'completed': 1, 'none': 0})
df_corr['lunch_num'] = df_corr['lunch'].map({'standard': 1, 'free/reduced': 0})

correlation_matrix = df_corr[['math score', 'reading score', 'writing score',
                             'test_prep_num', 'lunch_num']].corr()
print(correlation_matrix)
```

```python
print("=== PERFORMANCE GAPS ANALYSIS ===")

# Gender gap
gender_gap = gender_scores.loc['male'] - gender_scores.loc['female']
print("Gender Gap (Male - Female):")
print(gender_gap)

# Race/ethnicity range
race_range = race_scores.max() - race_scores.min()
print("\nRace/Ethnicity Score Range (Max - Min):")
print(race_range)

# Parental education impact range
parent_range = parent_scores.max() - parent_scores.min()
print("\nParental Education Impact Range (Max - Min):")
print(parent_range)
```

```python
print("=== KEY FINDINGS FOR STRATEGIC DECISIONS ===")

# 1. Where to focus improvement efforts
weakest_subject = avg_scores.idxmin()
print(f"1. Focus Area: {weakest_subject} has the lowest average score")

# 2. Most impactful intervention
prep_benefit = prep_impact.loc['completed'] - prep_impact.loc['none']
most_beneficial_subject = prep_benefit.idxmax()
print(f"2. Test Prep Impact: Most beneficial for {most_beneficial_subject}")

# 3. Equity focus
largest_gap_demographic = race_range.idxmax()
print(f"3. Equity Focus: Largest score gap in {largest_gap_demographic}")

# 4. Parental engagement opportunity
parent_impact = parent_range.idxmax()
print(f"4. Parental Engagement: Highest impact on {parent_impact}")
```

```python
plt.figure(figsize=(15, 10))

# Average scores by demographic factors
plt.subplot(2, 2, 1)
gender_scores.plot(kind='bar', ax=plt.gca())
plt.title('Average Scores by Gender')
plt.xticks(rotation=45)

plt.subplot(2, 2, 2)
race_scores.plot(kind='bar', ax=plt.gca())
plt.title('Average Scores by Race/Ethnicity')
plt.xticks(rotation=45)

plt.subplot(2, 2, 3)
parent_scores.sort_values('math score').plot(kind='bar', ax=plt.gca())
plt.title('Average Scores by Parental Education')
plt.xticks(rotation=45)

plt.subplot(2, 2, 4)
prep_impact.plot(kind='bar', ax=plt.gca())
plt.title('Test Preparation Impact')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
